```python
def get_dataframe(dataframe, patid_l, target_l):
    columns = ['s_entropy', 'entropy', 'moments', 'hu-moments', 'max_area', 'eccentricity', 'euler_number', 'solidity', 'perimeter', 'mean_area', 'std_area', 'thresh_img', 'correlation', 'homogeneity', 'contrast', 'energy']
    _moments = ['m00', 'm10', 'm01', 'm20', 'm11', 'm02', 'm30', 'm21', 'm12', 'm03', 'mu20', 'mu11', 'mu02', 'mu30', 'mu21', 'mu12', 'mu03', 'nu20', 'nu11',
                              'nu02', 'nu30', 'nu21', 'nu12', 'nu03']

    _hu = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7']

    _corr = ['corr1', 'corr2', 'corr3', 'corr4', 'corr5', 'corr6', 'corr7', 'corr8', 'corr9', 'corr10', 'corr11', 'corr12']

    _homo = ['homo1', 'homo2', 'homo3', 'homo4', 'homo5', 'homo6', 'homo7', 'homo8', 'homo9', 'homo10', 'homo11', 'homo12']


    _contrast = ['cont1', 'cont2', 'cont3', 'cont4', 'cont5', 'cont6', 'cont7', 'cont8', 'cont9', 'cont10', 'cont11', 'cont12']

    _energy = ['en1', 'en2', 'en3', 'en4', 'en5', 'en6', 'en7', 'en8', 'en9', 'en10', 'en11', 'en12']

    _drop_cols = ['moments', 'hu-moments', 'correlation', 'homogeneity', 'contrast', 'energy']
    
    df_moments = dataframe[2].apply(pd.Series)
    
#     df_moments = df_moments[1].apply(pd.Series)
    
    df_moments.columns = _moments
    
    df_hu = dataframe[3].apply(pd.Series)
    
    df_hu.columns = _hu
    
    df_corr = dataframe[12].apply(pd.Series)
    
    df_corr.columns = _corr
    
    df_homo = dataframe[13].apply(pd.Series)
    
    df_homo.columns = _homo
    
    df_contr = dataframe[14].apply(pd.Series)
    
    df_contr.columns = _contrast
    
    df_energy = dataframe[15].apply(pd.Series)
    
    df_energy.columns = _energy
    
    dataframe.columns = columns
    
    dataframe = dataframe.drop(_drop_cols, axis = 1)
    
    df_list = [dataframe, df_moments, df_hu, df_corr, df_homo, df_contr, df_energy]
    
    df_dcm_x = reduce(lambda x, y: pd.merge(x,y, right_index=True, left_index=True),df_list)
    
#     keys = df_dcm_x['s_entropy'].keys()
    
#     cls = []
    
#     for pid in keys:
#         if pid in patid_l:
#             idx = patid_l.index(pid)
#             cls.append(target[idx])
            
#     df_dcm_x['class'] = cls
    
    return df_dcm_x
```
